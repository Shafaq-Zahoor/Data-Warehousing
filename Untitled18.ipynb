```python
df_fact1 = pd.read_sql(''' SELECT Member_number, rfm_recency, rfm_frequency, rfm_monetary, 
rfm_recency*100 +rfm_frequency*10 + rfm_monetary AS rfm_combined
FROM
( SELECT Member_number,
NTILE(4) OVER (ORDER BY Date) AS rfm_recency,
NTILE(4) OVER (ORDER BY count_order) AS rfm_frequency,
NTILE(4) OVER (ORDER BY itemDescription) AS rfm_monetary
FROM
clean_data)'''
, conn)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def rfma(df_fact1):
  if ((df_fact1['rfm_recency'] >= 4) and (df_fact1['rfm_frequency'] >= 4)and (df_fact1['rfm_monetary'] >= 4)):
    return 'Best Customers'
  if ((df_fact1['rfm_recency'] >= 3) and (df_fact1['rfm_frequency'] >= 3)and (df_fact1['rfm_monetary'] >= 3)):
    return 'Loyal'
  if ((df_fact1['rfm_recency'] >= 3) and (df_fact1['rfm_frequency'] >= 1)and (df_fact1['rfm_monetary'] >= 2)):
    return 'Potential Loyalist'
  if ((df_fact1['rfm_recency'] >= 3) and (df_fact1['rfm_frequency'] >= 1)and (df_fact1['rfm_monetary'] >= 1)):
    return 'Promising'
  if ((df_fact1['rfm_recency'] >= 2) and (df_fact1['rfm_frequency'] >= 2)and (df_fact1['rfm_monetary'] >= 2)):
    return 'Customers Needing Attention'
  if ((df_fact1['rfm_recency'] >= 1) and (df_fact1['rfm_frequency'] >= 2)and (df_fact1['rfm_monetary'] >= 2)):
    return 'At Risk'
  if ((df_fact1['rfm_recency'] >= 1) and (df_fact1['rfm_frequency'] >= 1)and (df_fact1['rfm_monetary'] >= 2)):
    return 'Hibernating'
  else:
    return 'Lost'

df_fact1['rfma'] = df_fact1.apply(rfma, axis =1)
```

```python
# Calculate total customers in each segment
rfm_agg = df_fact1.groupby('rfma').agg({'Member_number':'count'})
print(rfm_agg)
```

```python
df_fact.dtypes
df_fact['last_order_date'] = pd.to_datetime(df_fact['last_order_date'])
df_fact.dtypes
# Create date variable that records recency
import numpy as np
import datetime
df_fact["last_order_date"] = df_fact["last_order_date"].dt.date
# Create snapshot date
snapshot_date = df_fact['last_order_date'].max() + datetime.timedelta(days=1)
print(snapshot_date)
```

```python
customers = df_fact.groupby(['Member_number']).agg({'last_order_date': lambda x: (snapshot_date - x.max()).days,'count_order':'sum' ,'totalitems': 'sum'})
```

```python
customers.rename(columns = {'last_order_date': 'Recency','count_order': 'Frequency','totalitems': 'MonetaryValue'}, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.histplot(data=customers, x='Recency', kde=True)
ax.set_title('Histogram and Recency')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot RFM distributions
plt.figure(figsize=(12,10))
# Plot distribution of R
plt.subplot(3, 1, 1); 
sns.distplot(customers['Recency'])
# Plot distribution of F
plt.subplot(3, 1, 2); 
sns.distplot(customers['Frequency'])
# Plot distribution of M
plt.subplot(3, 1, 3); 
sns.distplot(customers['MonetaryValue'])
# Show the plot
plt.show()
```

```python
from scipy import stats
customers_fix = pd.DataFrame()
customers_fix["Recency"] = stats.boxcox(customers['Recency'])[0]
customers_fix["Frequency"] = stats.boxcox(customers['Frequency'])[0]
customers_fix["MonetaryValue"] = stats.boxcox(customers['MonetaryValue'])[0]

customers_fix.tail()
```

```python
# Import library
from sklearn.preprocessing import StandardScaler
# Initialize the Object
scaler = StandardScaler()
# Fit and Transform The Data
scaler.fit(customers_fix)
customers_normalized = scaler.transform(customers_fix)
# Assert that it has mean 0 and variance 1
print(customers_normalized.mean(axis = 0).round(2)) # [0. -0. 0.]
print(customers_normalized.std(axis = 0).round(2)) # [1. 1. 1.]
```

```python
from sklearn.cluster import KMeans
sse = {}
for k in range(1, 11):
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(customers_normalized)
  sse[k] = kmeans.inertia_ # SSE to closest cluster centroid
  plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
model = KMeans(n_clusters=4, random_state=42)
model.fit(customers_normalized)
model.labels_.shape
customers["Cluster"] = model.labels_
customers.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    }).round(2)
f, ax = plt.subplots(figsize=(25, 5))
ax = sns.countplot(x="Cluster", data=customers)
customers.groupby(['Cluster']).count()
```

```python
customers["Cluster"] = model.labels_
customers.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'MonetaryValue' : 'mean'}).round(2)
```
